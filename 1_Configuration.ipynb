```python
# COMMAND ----------

# 1. Define widgets to accept parameters from the job run.
dbutils.widgets.text("catalog", "w_dev", "The Unity Catalog to use")
dbutils.widgets.text("landing_root", "s3a://wartsila-datalake-dev-landing/fingrid/", "The root S3 path for the landing zone")
dbutils.widgets.text("managed_uc_root", "s3a://wartsila-datalake-dev-managed-uc/", "The root S3 path for the managed Unity Catalog")
dbutils.widgets.text("bronze_root", "s3a://wartsila-datalake-dev-bronze/", "The root S3 path for the bronze layer")
dbutils.widgets.text("silver_root", "s3a://wartsila-datalake-dev-silver/", "The root S3 path for the silver layer")
dbutils.widgets.text("gold_root", "s3a://wartsila-datalake-dev-gold/", "The root S3 path for the gold layer")


# 2. Read the values from the widgets.
CATALOG = dbutils.widgets.get("catalog")
MANAGED_UC_ROOT = dbutils.widgets.get("managed_uc_root")
BRONZE_ROOT = dbutils.widgets.get("bronze_root")
SILVER_ROOT = dbutils.widgets.get("silver_root")
GOLD_ROOT = dbutils.widgets.get("gold_root")
LANDING_ROOT = dbutils.widgets.get("landing_root")


# 3. Build and execute the SQL commands.
print(f"Creating catalog '{CATALOG}' if it does not exist...")
spark.sql(f"CREATE CATALOG IF NOT EXISTS {CATALOG} MANAGED LOCATION '{MANAGED_UC_ROOT}'")

print(f"Creating schemas in catalog '{CATALOG}' if they do not exist...")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.bronze MANAGED LOCATION '{BRONZE_ROOT}'")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.silver MANAGED LOCATION '{SILVER_ROOT}'")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.gold MANAGED LOCATION '{GOLD_ROOT}'")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.landing_admin MANAGED LOCATION '{LANDING_ROOT}'")

print("Configuration complete.")

```
